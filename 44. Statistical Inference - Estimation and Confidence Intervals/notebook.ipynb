```python
import numpy as np
from scipy.stats import norm, t
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
import pandas as pd
```

### Sample

```python
# Sample Data
data = [12, 14, 15, 16, 17, 18, 19]

# Calculate Mean and Standard Deviation
mean = np.mean(data)
std = np.std(data, ddof=1) 

# 95% Confidence Interval using t-distribution
n = len(data)
t_value = t.ppf(0.975, df=n-1)
margin_of_error = t_value * (std / np.sqrt(n))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print(f"Mean: {mean}")
print(f"Standard Deviation: {std}")
print(f"95% Confidence Interval: {confidence_interval}")
```

### Exercise


```python
# Exercise 1: Calculate the confidence interval for sample data

# Generate random sample data
data = np.random.normal(loc=50, scale=10, size=100)

# Sample statistics
mean = np.mean(data)
std = np.std(data, ddof=1)

# 95% Confidence Interval
z_value = norm.ppf(0.975)
margin_of_error = z_value * (std / np.sqrt(len(data)))
ci = (mean - margin_of_error, mean + margin_of_error)
print(f"Sample Mean: {mean}")
print(f"Sample Standard Deviation: {std}")
print(f"95% Confidence Interval: {ci}")
```

```python
# Exercise 2: Conduct sampling and create a report

# Load the Iris dataset
df = pd.read_csv(url)

# Sampling
sample = df["sepal_length"].sample(n=30, random_state=42)

# Sample statistics
mean = sample.mean()
std = sample.std()
n = len(sample)

# 95% Confidence Interval using z-distribution
z_value = norm.ppf(0.975)
margin_of_error = z_value * (std / np.sqrt(n))
ci = (mean - margin_of_error, mean + margin_of_error)

print(f"Sample Mean: {mean}")
print(f"95% Confidence Interval: {ci}")
```
