```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

### Sample

```python
# Random Variable: dice roll
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6] * 6)

# Expected Value
expectation = np.sum(outcomes * probabilities)
print(f"Expected Value of a dice roll: {expectation}")

# Variance and Standard Deviation
variance = np.sum((outcomes - expectation) ** 2 * probabilities)
std_dev = np.sqrt(variance)
print(f"Variance of a dice roll: {variance}")
print(f"Standard Deviation of a dice roll: {std_dev}")
```

### Example


```python
# Simulate 10000 dice rolls
rolls = np.random.randint(1, 7, size=10000)

# Calculate probabilities
P_even = np.sum(rolls % 2 == 0) / len(rolls)
# P_odd = np.sum(rolls % 2 != 0) / len(rolls)
P_greater_than_4 = np.sum(rolls > 4) / len(rolls)

print ("P(Even):", P_even)
print("P(Greater than 4):", P_greater_than_4)
```

```python
# Discrete random variable: Dice roll
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
plt.bar(outcomes, probabilities, color='blue', alpha=0.7)
plt.title('Probability Mass Function of a Dice Roll')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()

# Continuous random variable: Uniform distribution
x = np.linspace(0, 1, 100)
pdf = uniform.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, color='red')
plt.title('Probability Density Function of Uniform(0, 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
