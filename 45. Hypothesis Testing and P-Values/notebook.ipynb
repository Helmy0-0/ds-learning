```python
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind
```

### Exercise 1: Perform a Hypothesis Test

```python
data = [12, 14, 15, 16, 17, 18, 19]

# Null Hypothesis: The mean of the sample data is equal to 15
population_mean = 15

# Perform t-test
t_stat, p_value = ttest_1samp(data, population_mean)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: no significant difference from the population mean.")
```
